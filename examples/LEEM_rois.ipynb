# LEEM - How to make ROIs and take stack profiles

For a basic tutorial for plotting intensity profiles, see `LEEM_plotting.ipynb`. This tutorial shows how to have multiple ROIs at the same time.

```python
from uspy import leem
```

A ROI is defined either as a circle, rectangle or ellipse. The first two arguments specify the center. If no specific type is given, a circle with a radius of 10px is assumed:

```python
roi1 = leem.ROI(300, 300)     # same as leem.ROI(x0=300, y0=300)
print(f"{roi1.type_=}")
print(f"{roi1.params=}")
```

If either parameters like `width` is given or an explicit `type_` (note the underscore), the ROI is different:

```python
roi2 = leem.ROI(300, 300, width=20, height=80)
print(f"{roi2.type_=}")

roi3 = leem.ROI(200, 350, xradius=40)
print(f"{roi3.type_=}")
print(f"{roi3.params=}")
```

In most plotting functions, you can either give a ROI, a list/tuple of ROIs or the arguments as used above:

```python
leem.plot_intensity_img("../testdata/test_stack_IV_g-Cu", x0=400, y0=400, radius=20)
```

```python
leem.plot_intensity_img("../testdata/test_stack_IV_g-Cu", (roi1, roi2, roi3))
```
