# How to calculate the dose of a gas in Langmuir

First, import the modules and make a stack. Refer to the tutorials "LEEM_plotting" and "LEEM_base" to understand what is going on here:

```python
from uspy import leem
from uspy.leem import processing
```

```python
stack = leem.LEEMStack("../testdata/test_stack_IV_g-Cu")
```

Now, calculate the dosage of some gas during the stack. By default "pressure1" is used to do this, but you can also select another field if you use `pressurefield=`:

```python
dose = processing.calculate_dose(stack, pressurefield="pressure2")
```

The easiest way to plot this now is to "monkey-patch" it to the stack and then use `plot_meta()`:

```python
stack.dose = dose
leem.plot_meta(stack, fields="dose")
```
